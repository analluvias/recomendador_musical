# A partir daqui vamos testar a obtenção das músicas similares a partir da letra

```python
# Carregar o modelo de linguagem em português do spaCy
nlp = spacy.load("pt_core_news_sm")
```

```python
def preprocessar_texto(text):
    doc = nlp(text)
    # pega a 'raíz' (forma base) da palavra e desconsidera pontos e stopwords
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    return ' '.join(tokens)

df_merged['texto_processado'] = df_merged['letra'].apply(preprocessar_texto)
```

```python
# Iterate over the rows of the DataFrame and print the 'texto_expandido' column
for index, row in df_merged.iterrows():
  pass
  # print(row['texto_processado'])
```

```python
# Calcular o TF-IDF
vectorizer = TfidfVectorizer(min_df=2, max_df=0.7)
tfidf_matrix = vectorizer.fit_transform(df_merged['texto_processado'])

# Exibir a matriz TF-IDF
print(tfidf_matrix.toarray())
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(),columns=vectorizer.get_feature_names_out())
tfidf_df.index = df_merged['Música']
print(tfidf_df)
```

```python
# Import cosine_similarity measure
from sklearn.metrics.pairwise import cosine_similarity

# Create the array of cosine similarity values
cosine_similarity_array = cosine_similarity(tfidf_df)

# Wrap the array in a pandas DataFrame
cosine_similarity_df = pd.DataFrame(cosine_similarity_array, index=tfidf_df.index, columns=tfidf_df.index)

# Print the top 5 rows of the DataFrame
cosine_similarity_df.head()
```

# Função que retorna as 5 músicas mais parecidas a partir da similaridade do cosseno

```python
# Wrap the preloaded array in a DataFrame
cosine_similarity_df = pd.DataFrame(cosine_similarity_array, index=tfidf_df.index, columns=tfidf_df.index)
```

```python
def find_5_similar_songs(song_title):
  # Find the values for the music
  cosine_similarity_series = cosine_similarity_df[song_title]

  # Sort these values highest to lowest
  ordered_similarities = cosine_similarity_series.sort_values(ascending=False)

  # Merge with df_merged to get the 'Tema_y' column
  ordered_similarities = pd.merge(ordered_similarities, df_merged[['Música', 'Tema_y']], left_index=True, right_on='Música')

  return ordered_similarities[:5]
```

```python
import os
import glob
import opensmile

def extract_all_wav_feature(wavdir, distfile):
    smile = opensmile.Smile(
        feature_set=opensmile.FeatureSet.ComParE_2016,
        feature_level=opensmile.FeatureLevel.Functionals,
    )

    all_features = []
    for wavpath in glob.glob(os.path.join(wavdir, '*.wav')):
        features = smile.process_file(wavpath)
        features['file'] = os.path.basename(wavpath)
        all_features.append(features)

    df = pd.concat(all_features)
    df.to_csv(distfile, index=False)

def extract_frame_feature(wavdir, distdir):
    smile = opensmile.Smile(
        feature_set=opensmile.FeatureSet.ComParE_2016,
        feature_level=opensmile.FeatureLevel.LowLevelDescriptors,
    )

    if os.path.exists(distdir):
        os.system(f"rm -rf {distdir}")
    os.makedirs(distdir)

    for wavpath in glob.glob(os.path.join(wavdir, '*.wav')):
        features = smile.process_file(wavpath)
        filename = os.path.basename(wavpath).replace('.wav', '.csv')
        distpath = os.path.join(distdir, filename)
        features.to_csv(distpath, index=False)

def process_dynamic_feature(llddir, distdir, all_songs_distfile):
    if os.path.exists(distdir):
        os.system(f"rm -rf {distdir}")
    os.makedirs(distdir)

    all_features = []
    for lldpath in glob.glob(os.path.join(llddir, '*.csv')):
        df = pd.read_csv(lldpath)
        df['file'] = os.path.basename(lldpath).replace('.csv', '')
        all_features.append(df)

    combined_df = pd.concat(all_features)
    combined_df.to_csv(all_songs_distfile, index=False)

# Bloco principal
if __name__ == "__main__":
    wavdir = '/content/drive/MyDrive/projeto-metodos-numericos/musicas_youtube'
    distfile = '/content/drive/MyDrive/projeto-metodos-numericos/dados_nossas_musicas/primeiro.csv'

    extract_all_wav_feature(wavdir, distfile)
    extract_frame_feature(wavdir, '/content/drive/MyDrive/projeto-metodos-numericos/dados_nossas_musicas/individual')

    process_dynamic_feature('/content/drive/MyDrive/projeto-metodos-numericos/dados_nossas_musicas/individual',
                            '/content/drive/MyDrive/projeto-metodos-numericos/dados_nossas_musicas/pasta_all_concat',
                            '/content/drive/MyDrive/projeto-metodos-numericos/dados_nossas_musicas/pasta_all_concat/all_songs.csv')
```

O LinearSVR (Linear Support Vector Regression) é um algoritmo de regressão que utiliza um kernel linear para prever valores contínuos. Ele é uma variante do Support Vector Regression (SVR) que se baseia em conceitos fundamentais de álgebra linear. Vamos explorar essa correlação:
Função de Decisão Linear
A função de decisão do LinearSVR é uma combinação linear das características de entrada ( x ):
f(x)=w⋅x+b
onde ( w ) é o vetor de pesos (coeficientes) e ( b ) é o termo de intercepto. Esta é uma expressão típica de uma reta em um espaço multidimensional, que é um conceito central em álgebra linear.
Problema de Otimização
O treinamento do LinearSVR envolve a resolução de um problema de otimização que minimiza uma função de perda regularizada. A função de perda epsilon-insensitive é usada para ignorar erros menores que um valor epsilon:
w,bmin​21​∥w∥2+Ci=1∑n​max(0,∣yi​−(w⋅xi​+b)∣−ϵ)
Aqui, ( C ) é um parâmetro de regularização que controla o trade-off entre a margem de erro e a complexidade do modelo1.
Vetores de Suporte
Assim como no SVM, o LinearSVR utiliza vetores de suporte, que são os pontos de dados mais próximos à linha de regressão. Esses pontos são críticos para definir a posição e a inclinação da linha de regressão.
Regularização
A regularização é uma técnica usada para evitar o overfitting, adicionando um termo de penalização à função de perda. No caso do LinearSVR, a regularização é aplicada aos coeficientes ( w ), o que é essencialmente uma forma de controle sobre a norma ( |w| ).

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVR
from sklearn.pipeline import make_pipeline

# Carregar os dados
data_a = load_static_features_and_valence()
features_a, labels_a = load_audio_dataset(data_a)

# Definir o pré-processador e o modelo LinearSVR
prefunc = [StandardScaler()]
regressor = LinearSVR()
model = make_pipeline(*prefunc, regressor)

# Treinar o modelo com todos os dados
model.fit(features_a, labels_a)
```
